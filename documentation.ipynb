Team: Yvonne Gisser, Dusan Resavac, Roland Bauer

## Einleitung

### Kurzbeschreibung
Wir möchten Filmrezensionen auf eine einfache und anschauliche Art zusammenfassen, sodass der Konsens verschiedener Meinungen auf einem Blick ersichtlich ist. Dies soll durch die Darstellung der am häufigsten vorkommenden Adjektive geschehen. Dafür möchten wir eine Word Cloud (Schlagwortwolke) zu beliebigen Filmen erstellen können, wobei die Größe des Wortes proportional zur Häufigkeit steht.

### Geplante Datenspeicherung

Wir haben geplant die Daten zumindest einer Datenquelle in einer NoSQL-Datenbank zu speichern. Beispielsweise könnten Wörter, die zur Gruppe der Adjektive gehören, in einer Redis-DB gespeichert werden. Dabei wären die Keys die Wörter und die Values irrelevant. In einer anderen Datenbank würden wir die Rezensionstexte und einige zugehörige Filmdaten speichern.

### Geplante Vorgehensweise

Zuerst würden wir die Wörter in einer Datenbank speichern (Entweder alle + Wortart oder nur Adjektive). Danach können ausgewählte Filme über einen der vorher erwähnten Endpunkte geladen werden. Zu den Filmen werden dann Rezensionen geladen, die mit Kafka über ein Topic bereitgestellt werden. Der Consumer holt sich die Rezensionstexte und lädt das Wörterbuch. Dann erfolgt nach dem MapReduce Prinzip mit Spark das Berechnen der Anzahl der einzelnen Adjektive. Das Ergebnis wird dann gespeichert und als Visualisierung wird eine Word Cloud erstellt. ### Erwarteter Output

Als Output erwarten wir eine Word Cloud, die zum Beispiel auf die 5 am häufigsten vorkommenden Wörter eingeschränkt ist. Die Größe jedes einzelnen Wortes wird von der relativen Häufigkeit bestimmt. Je nach Umsetzbarkeit kann auf eine leichter umsetzbare Variante ausgewichen werden. Beispielhafte Outputs sind am Anfang vom Dokument zu sehen.
